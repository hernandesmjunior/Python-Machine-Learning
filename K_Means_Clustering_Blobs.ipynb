# K Means Clustering

## Importando as bibliotecas

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8, random_state=50)
```

```python
type(data)
```

```python
data
```

```python
plt.figure(figsize=(12,6))
plt.scatter(data[0][:,0],data[0][:,1],c=data[1], cmap='inferno')
```

## Modelo de clusterizao

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1,2,sharey=True, figsize=(12,8))
ax1.set_title('KMeans')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_, cmap='inferno')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1], cmap='inferno')
```
